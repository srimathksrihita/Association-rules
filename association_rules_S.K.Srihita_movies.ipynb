```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
movies=pd.read_csv('C:\\Users\\user\\Downloads\\my_movies.csv')
```

```python
movies
```

```python
movies.shape
```

```python
movies.info()
```

```python
df=movies.iloc[:,5:]
```

```python
df
```

```python
#apriori algoritm
```

```python
#10% support and 70%confidence
```

```python
frequent_itemsets=apriori(df,min_support=0.1,use_colnames=True)
```

```python
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=0.7)
```

```python
rules
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
#support is 20 % confidence 60 %
```
